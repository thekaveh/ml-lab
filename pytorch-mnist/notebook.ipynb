```python
import numpy as np
import torchvision as thv

from torch import torch
from torch.utils.data import DataLoader
from torch.utils.data.sampler import SubsetRandomSampler, SequentialSampler

from utils import Utils
from consts import Consts
from feed_fwd_nn_model import FeedFwdNNModel
```

```python
data_train = thv.datasets.MNIST(
    root="./data"
    , train=True
    , download=True
    , transform=thv.transforms.ToTensor()
)

I_train = np.random.randint(
    low=0
    , high=data_train.data.shape[0]
    , size=int(data_train.data.shape[0] * Consts.DS_SAMPLING_RATIO)
)

loader_train = DataLoader(
    dataset=data_train
    , batch_size=Consts.MINI_BATCH_SIZE
    , sampler=SubsetRandomSampler(I_train)
)

print(I_train.shape)
print(len(loader_train.sampler))
```

```python
data_val = thv.datasets.MNIST(
    root="./data"
    , train=False
    , download=True
    , transform=thv.transforms.ToTensor()
)

I_val = np.arange(0, int(data_val.data.shape[0] * Consts.DS_SAMPLING_RATIO))

loader_val = DataLoader(
    dataset=data_val
    , batch_size=Consts.MINI_BATCH_SIZE
    , sampler=SequentialSampler(I_val)
)

print(I_val.shape)
print(len(loader_val.sampler))
```

```python
model1 = FeedFwdNNModel(
    lr=0.1
    , dropout_p=0.2
    , optimizer_alg="sgd"
    , loader_val=loader_val
    , loader_train=loader_train
    , device_name=Utils.get_device_name()
)

model1_idps = model1.train_and_validate()
```

```python
Utils.two_line_plot(
    y1_legend="Training"
    , y_axis_label="Loss"
    , y2_legend="Validation"
    , x_axis_label="Iterations"
    , fig_size=(2 * 2 * 2 * 3, 2 * 2 * 3)
    , x=[idp.iter_idx for idp in model1_idps]
    , y1=[idp.training_loss for idp in model1_idps]
    , y2=[idp.validation_loss for idp in model1_idps]
    , title=f"Training & Validation Losses for {model1}"
)
```

```python
Utils.two_line_plot(
    y1_legend="Training"
    , y_axis_label="Error"
    , y2_legend="Validation"
    , x_axis_label="Iterations"
    , fig_size=(2 * 2 * 2 * 3, 2 * 2 * 3)
    , x=[idp.iter_idx for idp in model1_idps]
    , y1=[idp.training_error for idp in model1_idps]
    , y2=[idp.validation_error for idp in model1_idps]
    , title=f"Training & Validation Errors for {model1}"
)
```
